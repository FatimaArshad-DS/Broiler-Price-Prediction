```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.agbro.com/'
data = requests.get(url)
```

```python
soup = BeautifulSoup(data.text, 'html.parser')
table = soup.find("table", attrs={"class":"tableizer-table"})
```

```python
# The first tr contains the field names.
date = []
day = []
DOC = []
fr = []
open = []
close = []
for row in table.find_all("tr")[3:]:
    td = row.find_all("td")[0:6]
    date.append(td[0].text)
    day.append(td[1].text)
    DOC.append(td[2].text)
    fr.append(td[3].text)
    open.append(td[4].text)
    close.append(td[5].text)
```

```python
df = pd.DataFrame()
df["date"] = date
df["day"] = day
df["doc"] = DOC
df["farm rate"] = fr
df["open"] = open
df["close"] = close
```

```python
years = ['2022','2021','2020','2019','2018','2017','2016','2015']
date = []
day = []
DOC = []
fr = []
open = []
close = []
for i in range(len(years)):
  url = 'https://www.agbro.com/broiler-market-prices-'+years[i]
  data = requests.get(url)
  soup = BeautifulSoup(data.text, 'html.parser')
  table = soup.find_all("table", attrs={"class":"tableizer-table"})
  for j in table:
    for row in j.find_all("tr")[3:]:
      td = row.find_all("td")[0:6]
      try:
        date.append(td[0].text)
      except:
        date.append("-")
        pass
      try:
        day.append(td[1].text)
      except:
        day.append("-")
        pass
      try:
        DOC.append(td[2].text)
      except:
        DOC.append("-")
        pass
      try:
        fr.append(td[3].text)
      except:
        fr.append("-")
        pass
      try:
        open.append(td[4].text)
      except:
        open.append("-")
        pass
      try:
        close.append(td[5].text)
      except:
        close.append("-")
        pass

```

```python
df = pd.DataFrame()
df["date"] = date
df["day"] = day
df["doc"] = DOC
df["farm rate"] = fr
df["open"] = open
df["close"] = close
```
